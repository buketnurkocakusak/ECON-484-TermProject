```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


#Excel is loaded. Header 0 is applied to get rid of first row.
dataset = pd.read_excel('EVDS (14).xlsx',header=0)
```

```
#Dataset obtained is checked to make sure whether the numbers loaded correctly.
dataset
```

```
#x values are gathered from the excel first column. The data was from 1 to 24th rows. That is why it is limited as "0:25".
x_data = np.array(dataset['Tarih'][0:25])
#y values are gathered from the excel second column. The data was from 1 to 24th rows. That is why it is limited as "0:25".
y_data = np.array(dataset['TP DK USD A YTL'][0:25])
#As the date data was large enough, it was necessary to extend the size of the grapth.
plt.figure(figsize=(25,10))
#Graph dots were combined with straight line.
plt.plot(x_data,y_data,'o-')
#Graph x and y values are named as it supposed to be.
plt.ylabel('TP DK USD A YTL')
plt.xlabel('Tarih')
#Background grid applied.
plt.grid()
plt.show()
```
